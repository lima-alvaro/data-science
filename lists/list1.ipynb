```python
import pandas as pd
```

#### a) Ler o dataset fakeTelegram.BR_2022.csv

```python
df = pd.read_csv("../data/fakeTelegram.BR_2022.csv")
df.head()
```

b) Identificar e listar as posições (células) contendo valores faltantes.


```python
null_cells = [
    f"Valor nulo na linha {i} e coluna {j}"
    for i in range(df.shape[0])
    for j in range(df.shape[1])
    if pd.isnull(df.iat[i, j])
]

print("Listando apenas as 20 primeiras células nulas (pois há muitos valores)")
print(*null_cells[:20], sep ='\n')
```

#### c) Contar quantas linhas possuem valores faltantes.


```python
total_rows = df.shape[0]
rows_without_nulls = df.dropna().shape[0]
total_rows - rows_without_nulls
```

#### d) Para cada coluna (feature), contar quantas linhas possuem valores faltantes.

```python
null_rows = df.isna()
null_rows.sum()
```

#### e) Identificar e listar as linhas repetidas (duplicadas).

```python
duplicated = df.duplicated(
    subset=["date_message", "text_content_anonymous", "id_member_anonymous", "id_group_anonymous", "id_message"]
)
duplicated.sum()
```

```python
duplicated_rows = df[duplicated]
duplicated_rows
```

#### f) Identificar e listar as posições (células) contendo valores que não pertencem ao domínio (tipo de dados) esperado.


#### g) Crie uma coluna chamada “caracteres” contendo a quantidade de caracteres da coluna “text”.


```python
df["caracteres"] = df["text_content_anonymous"].apply(lambda x: 0 if pd.isnull(x) else len(x))
df[["text_content_anonymous", "caracteres"]].head(10)
```

#### h) Crie uma coluna chamada “words” contendo a quantidade de palavras da coluna “text”.

```python
df["words"] = df["text_content_anonymous"].apply(lambda x: 0 if pd.isnull(x) else len(x.split()))
df[["text_content_anonymous", "words"]].head(10)
```

 #### i) Crie uma coluna chamada “viral” contendo o valor 0 se o texto da mensagem (valor do campo “text”) não for encontrado em outras linhas e 1, caso contrário.


```python
text_frequency = df["text_content_anonymous"].value_counts()
viral_texts = (text_frequency [text_frequency > 1]).index
df["viral"] = df["text_content_anonymous"].isin(viral_texts)
df[["text_content_anonymous", "viral"]].head(10)
```

#### j) Crie uma coluna chamada “sharings” contendo a quantidade de vezes que o texto armazenado no atributo “text” aparece no dataset.

```python
df["sharings"] = df["text_content_anonymous"].map(text_frequency)
df[["text_content_anonymous", "sharings"]].head(10)
```

#### k) Crie uma coluna chamada “sentiment” contendo os valores: -1 para textos negativos, 0 para textos neutros e 1 para textos positivos

```python
df["sentiment"] = df["score_sentiment"].apply(
    lambda x: 1 if x >= 0.05
    else -1 if x <= -0.05
    else 0
)

df[["text_content_anonymous", "sentiment"]].head(10)
```

#### l) Eliminar as linhas cujo valor da coluna “text” contenham “trava-zaps”.


```python
df = df[~df["trava_zap"]]
df.head(10)
```

#### m) Identificar inconsistências entre os atributos (features).

```python
distinct_types = [
    (column, df[column].map(type).unique())
    for column in df.columns
]

for column, types in distinct_types:
    print(f"Coluna '{column}', Tipos distintos: {types}")
```
