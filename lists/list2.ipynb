#### 21. Os 30 unigramas, bigramas e trigramas mais compartilhados;

unigramas

```python
con.execute(
'''
SELECT
    token AS unigrams,
    COUNT(*) AS count
FROM (
    SELECT UNNEST(string_split(lower(text_content_anonymous), ' ')) AS token
    FROM mensagens
    WHERE has_media = false AND has_media_url = false AND text_content_anonymous IS NOT NULL
) AS tokens
GROUP BY unigrams
ORDER BY count DESC
LIMIT 30
'''
).fetchdf()
```

bigramas

```python
con.execute(
'''
SELECT
    CONCAT_WS(' ', prev_token, token) AS bigram,
    COUNT(*) AS count
FROM (
    SELECT
        token,
        LAG(token) OVER (PARTITION BY message_id ORDER BY token_index) AS prev_token
    FROM (
        SELECT
            ROW_NUMBER() OVER () AS message_id,
            UNNEST(string_split(lower(text_content_anonymous), ' ')) AS token,
            ROW_NUMBER() OVER (PARTITION BY text_content_anonymous ORDER BY text_content_anonymous) AS token_index
        FROM mensagens
        WHERE has_media = false AND has_media_url = false AND text_content_anonymous IS NOT NULL
    ) AS tokenized
) AS bigrams
WHERE prev_token IS NOT NULL
GROUP BY bigram
ORDER BY count DESC
LIMIT 30
'''
).fetchdf()
```

trigramas

```python
con.execute(
'''
SELECT
    CONCAT_WS(' ', prev_token, token, next_token) AS trigram,
    COUNT(*) AS count
FROM (
    SELECT
        token,
        LAG(token) OVER (PARTITION BY message_id ORDER BY token_index) AS prev_token,
        LEAD(token) OVER (PARTITION BY message_id ORDER BY token_index) AS next_token
    FROM (
        SELECT
            ROW_NUMBER() OVER () AS message_id,
            UNNEST(string_split(lower(text_content_anonymous), ' ')) AS token,
            ROW_NUMBER() OVER (PARTITION BY text_content_anonymous ORDER BY text_content_anonymous) AS token_index
        FROM mensagens
        WHERE has_media = false AND has_media_url = false AND text_content_anonymous IS NOT NULL
    ) AS tokenized
) AS trigrams
WHERE prev_token IS NOT NULL AND next_token IS NOT NULL
GROUP BY trigram
ORDER BY count DESC
LIMIT 30
'''
).fetchdf()
```

#### 28. O dia em que foi publicado a maior quantidade de mensagens;


```python
con.execute(
'''
SELECT
    DATE_TRUNC('day', STRPTIME(date_message, '%Y-%m-%d %H:%M:%S')) AS dia,
    COUNT(*) AS quantidade_mensagens
FROM mensagens
GROUP BY dia
ORDER BY quantidade_mensagens DESC
LIMIT 1
'''
).fetchdf()
```
